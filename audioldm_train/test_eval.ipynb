```python
import os
from audioldm_train.eval import locate_validation_output


def find_evaluation_results(exp_folder):
    """
    exp_folder: 실험 폴더 경로
    평가 결과 JSON 파일의 위치를 찾아서 반환
    """
    # 실험 폴더의 상위 디렉토리 확인
    parent_dir = os.path.dirname(exp_folder)
    results = []
    
    # JSON 파일 찾기
    for file in os.listdir(parent_dir):
        if file.endswith('.json'):
            # 실험 폴더명이 JSON 파일명에 포함되어 있는지 확인
            if os.path.basename(exp_folder) in file:
                results.append(os.path.join(parent_dir, file))
    
    return results

# 사용 예시:
validation_folders = locate_validation_output("/home/summer/AudioLDM-training-finetuning/log/latent_diffusion/2023_08_23_reproduce_audioldm/audioldm_original_audio")
for folder in validation_folders:
    result_files = find_evaluation_results(folder)
    if result_files:
        print(f"Evaluation results for {folder}:")
        for result_file in result_files:
            print(f"- {result_file}")
    else:
        print(f"No evaluation results found for {folder}")
```
